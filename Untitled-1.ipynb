```python
import time
from genetic import *

def get_time(n):
    debt = [0] * n

    start = time.time()

    GENERATIONS = 500

    population = getInitialPopulation(n, debt)

    for _ in range(GENERATIONS):
        population = getNextGen(population, debt)
    
    end = time.time()

    return end - start
```

```python
import dynamic_programming as dp
import time
def get_dp_time(n):
    debt = [0] * n
    start = time.time()
    dp.main(debt)
    end = time.time()

    return end - start
```

```python
y = [get_time(i) for i in range(2, 100)]
z = [get_dp_time(i) for i in range(2, 20)]
```

```python
import matplotlib.pyplot as plt

plt.plot(z, label = "DP")

plt.plot(y, label = "GA")

plt.legend()

plt.xlabel("n")
plt.ylabel("time")

plt.show()
```
